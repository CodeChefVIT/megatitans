```
import pandas as pd
dataset = pd.read_csv('gdrive/My Drive/Colab/ebola_data_db_format.csv')
dataset['UID'] = dataset[['Country', 'Date']].apply(lambda x: ''.join(x), axis=1)
cols = [x for x in set(dataset['Indicator'])]
cols.append("Country")
cols.append("Date")
cols.append("UID")
df = pd.DataFrame(columns=cols)
```

```
for index, row in dataset.iterrows():
    try:
        r_no = df.loc[df["UID"] == row["UID"]].index.values.astype(int)[0]
        df.loc[r_no][row["Indicator"]] = row["value"]
        df.loc[r_no]["Country"] = row["Country"]
        df.loc[r_no]["Date"] = row["Date"]
    except:
        df = df.append({str(row["Indicator"]): row["value"], str(df['UID']) : row["UID"], "Date" : row["Date"], "Country" : row["Country"]}, ignore_index=True) 
 
```
